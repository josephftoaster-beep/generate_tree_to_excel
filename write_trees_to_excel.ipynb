```python
import pandas as pd
import os

def write_trees_to_excel(folder_data, output_filepath='Project_Tree_Report.xlsx'):
    """
    ãƒ•ã‚©ãƒ«ãƒ€å:ãƒ„ãƒªãƒ¼ãƒªã‚¹ãƒˆã®è¾æ›¸ã‚’å—ã‘å–ã‚Šã€å„ãƒ•ã‚©ãƒ«ãƒ€åã‚’ã‚·ãƒ¼ãƒˆåã¨ã—ã¦Excelã«æ›¸ãå‡ºã™é–¢æ•°ã€‚
    
    Args:
        folder_data (dict): generate_visual_tree é–¢æ•°ã‹ã‚‰è¿”ã•ã‚ŒãŸè¾æ›¸ã€‚
        output_filepath (str): å‡ºåŠ›ã™ã‚‹Excelãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ‘ã‚¹ã¨åå‰ã€‚
    """
    print(f'ğŸ“Š Excelãƒ•ã‚¡ã‚¤ãƒ«ã¸ã®æ›¸ãå‡ºã—ã‚’é–‹å§‹ã—ã¾ã™: {output_filepath}')
    
    # ç¾å ´ã®ä¾‹ãˆï¼šExcelãƒ–ãƒƒã‚¯ï¼ˆå°å¸³ï¼‰ã®ä½œæˆæº–å‚™
try:
        #ExcelWriterã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã‚’ä½œæˆã€‚withæ§‹æ–‡ã§ç¢ºå®Ÿã«ä¿å­˜ãƒ»ã‚¯ãƒ­ãƒ¼ã‚º
        with pd.ExcelWriter(output_filepath, engine='xlsxwriter') as writer:
            #foleder_data(è¾æ›¸)ã®ã‚­ãƒ¼ã¨å€¤ã‚’ãã‚Œãã‚Œfoleder_nameã¨tree_linesã«æ ¼ç´ã€‚
            for folder_name, tree_lines in folder_data.items():
                #ã‚¨ãƒ©ãƒ¼æƒ…å ±(è¿”ã‚Šå€¤)ã®ãƒã‚§ãƒƒã‚¯ã™ã‚‹ã€‚
                if folder_name == 'ã‚¨ãƒ©ãƒ¼':
                    print(f"è­¦å‘Šï¼šã‚¨ãƒ©ãƒ¼æƒ…å ±ã‚’Excelã«å«ã‚ãšã‚¹ã‚­ãƒƒãƒ—ã—ã¾ã—ãŸ: {tree_lines[0]}")
                    continue
                #ãƒ„ãƒªãƒ¼ãƒªã‚¹ãƒˆã‚’DataFrame(è¡¨å½¢å¼)ã«å¤‰æ›
                df = pd.DataFrame(tree_lines, columns=['Path']) 
                
                sheet_name = folder_name.replace(os.sep, '_').replace('/', '_').replace('\\', '_')
                sheet_name = sheet_name.strip('_')
                
                if not sheet_name: 
                    sheet_name = 'ãƒ«ãƒ¼ãƒˆ'
                
                sheet_name = sheet_name[:31] 
                
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                
                try:
                    worksheet = writer.sheets[sheet_name]
                    worksheet.set_column(0, 0, 50) 
                except Exception as e:
                    pass

        print(f"âœ… Excelãƒ•ã‚¡ã‚¤ãƒ«ãŒæ­£å¸¸ã«ä½œæˆã•ã‚Œã¾ã—ãŸ: {output_filepath}")

    except Exception as e:
        print(f"âŒ Excelãƒ•ã‚¡ã‚¤ãƒ«æ›¸ãå‡ºã—ä¸­ã«è‡´å‘½çš„ãªã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸ: {e}")
```
